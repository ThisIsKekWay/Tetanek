```python
import pandas as pd
```

```python
df = pd.read_csv('train (2).csv')
```

```python
df
```

```python
def cat(age):
    res = age.copy()
    for i in range(len(age)):
        if age[i] < 18:
            res[i] = 'Дети'
        elif 18 <= age[i] <= 30:
            res[i] = 'Молодые'
        elif 30 < age[i] <= 50:
            res[i] = 'Взрослые'
        elif 50 < age[i]:
            res[i] = 'Престарелые'
        else:
            res[i] = 'Неизвестно'
    return res
```

```python
df['category'] = cat(df.Age)
```

```python
df
```

```python
# Выживаемость пассажиров по категориям
df.groupby(['category']).Survived.value_counts(normalize = True).mul(100).round(2)
```
